```python
# =============================================================================
# Глава 1. Первичный анализ набора табличных данных
# Датасет: NHL Player Database (игроки НХЛ)
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# -----------------------------------------------------------------------------
# ФИОЛЕТОВАЯ ЦВЕТОВАЯ СХЕМА
# -----------------------------------------------------------------------------
PURPLE = '#9b59b6'      # основной фиолетовый
DARK_PURPLE = '#8e44ad' # темно-фиолетовый
LIGHT_PURPLE = '#c39bd3' # светлый фиолетовый
PALE_PURPLE = '#d2b4de'  # бледно-фиолетовый

# Настройка стилей
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

# Создаем папку для сохранения графиков
import os
if not os.path.exists('figures'):
    os.makedirs('figures')


# =============================================================================
# 1. ОБЩЕЕ ОПИСАНИЕ ДАТАСЕТА
# =============================================================================

# Загрузка данных
skaters = pd.read_csv('SKATERS.csv')
goalies = pd.read_csv('GOALIES.csv')

# Добавляем признак "очки" = голы + передачи
skaters['points'] = skaters['goals'] + skaters['assists']
goalies['points'] = goalies['goals'] + goalies['assists']

print("=" * 60)
print("1. ОБЩЕЕ ОПИСАНИЕ ДАТАСЕТА")
print("=" * 60)

print("\n--- ИСТОЧНИК ДАННЫХ ---")
print("Название: NHL Player Database")
print("Источник: https://www.kaggle.com/datasets/flynn28/nhl-player-database")
print("Лицензия: открытая, для образовательных целей")

print("\n--- SKATERS.csv (полевые игроки) ---")
print(f"Количество игроков: {len(skaters)}")
print(f"Количество признаков: {len(skaters.columns)}")
print(f"\nПризнаки:")
for col in skaters.columns:
    print(f"  - {col}: {skaters[col].dtype}")

print("\n--- GOALIES.csv (вратари) ---")
print(f"Количество вратарей: {len(goalies)}")
print(f"Количество признаков: {len(goalies.columns)}")

print("\n--- ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ ---")
print(f"В SKATERS пропущены в 'time': {skaters['time'].isnull().sum()} записей ({skaters['time'].isnull().sum()/len(skaters)*100:.1f}%)")
print(f"В GOALIES пропущены в разных полях: есть (до 60% в некоторых)")

print("\n--- КОНФИДЕНЦИАЛЬНОСТЬ ---")
print("Данные являются спортивной статистикой, не содержат персональных данных. Все игроки публичные личности.")


# =============================================================================
# 2. ХАРАКТЕРИСТИКА ЧИСЛОВЫХ ДАННЫХ
# =============================================================================

print("\n" + "=" * 60)
print("2. ХАРАКТЕРИСТИКА ЧИСЛОВЫХ ДАННЫХ")
print("=" * 60)

# Выбираем признаки для анализа
numeric_cols = ['games', 'goals', 'assists', 'points']
print(f"\nЧисловые признаки: {numeric_cols}")


# -----------------------------------------------------------------------------
# 2.1 ДИАГРАММЫ РАСПРЕДЕЛЕНИЯ (гистограммы)
# -----------------------------------------------------------------------------

print("\n--- 2.1 Диаграммы распределения ---")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(skaters[col].dropna(), bins=40,
                 color=PURPLE, edgecolor=DARK_PURPLE, alpha=0.7)
    axes[i].set_title(f'Распределение: {col}', fontsize=14)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Количество игроков')

plt.tight_layout()
plt.savefig('figures/1_distributions.png', dpi=150, bbox_inches='tight')
plt.show()

print("Вывод: Все признаки имеют правостороннее распределение.")
print("Большинство игроков имеют средние показатели, но есть звезды с высокими значениями.")


# -----------------------------------------------------------------------------
# 2.2 ВИЗУАЛИЗАЦИЯ ПАР ПРИЗНАКОВ (Seaborn)
# -----------------------------------------------------------------------------

print("\n--- 2.2 Визуализация пар признаков ---")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# График 1: голы vs передачи
sns.scatterplot(data=skaters, x='goals', y='assists',
                color=PURPLE, alpha=0.5, ax=axes[0])
axes[0].set_title('Голы vs Передачи', fontsize=14)
axes[0].set_xlabel('Голы')
axes[0].set_ylabel('Передачи')

# График 2: игры vs очки
sns.scatterplot(data=skaters, x='games', y='points',
                color=PURPLE, alpha=0.5, ax=axes[1])
axes[1].set_title('Игры vs Очки', fontsize=14)
axes[1].set_xlabel('Количество игр')
axes[1].set_ylabel('Очки')

# График 3: голы vs очки
sns.scatterplot(data=skaters, x='goals', y='points',
                color=PURPLE, alpha=0.5, ax=axes[2])
axes[2].set_title('Голы vs Очки', fontsize=14)
axes[2].set_xlabel('Голы')
axes[2].set_ylabel('Очки')

plt.tight_layout()
plt.savefig('figures/2_scatter_pairs.png', dpi=150, bbox_inches='tight')
plt.show()

print("Вывод: Сильная корреляция между голами и передачами, а также между играми и очками.")


# -----------------------------------------------------------------------------
# 2.3 ИНТЕРАКТИВНЫЕ ГРАФИКИ (Plotly)
# -----------------------------------------------------------------------------

print("\n--- 2.3 Интерактивные графики ---")

# Интерактивная гистограмма голов
fig1 = px.histogram(skaters, x='goals', nbins=50,
                    title='Распределение голов (интерактивный)',
                    color_discrete_sequence=[PURPLE])
fig1.show()

# Интерактивный точечный график
fig2 = px.scatter(skaters, x='games', y='points',
                  title='Очки vs Игры (интерактивный)',
                  labels={'games': 'Количество игр', 'points': 'Очки'},
                  color_discrete_sequence=[PURPLE])
fig2.show()


# -----------------------------------------------------------------------------
# 2.4 ОБРАБОТКА ПРОПУСКОВ
# -----------------------------------------------------------------------------

print("\n--- 2.4 Обработка пропусков ---")

# Заполняем пропуски в time средним значением
time_mean = skaters['time'].mean()
skaters['time'].fillna(time_mean, inplace=True)
print(f"Пропуски в 'time' заполнены средним: {time_mean:.0f} минут")


# -----------------------------------------------------------------------------
# 2.5 ТЕПЛОВАЯ КАРТА КОРРЕЛЯЦИЙ
# -----------------------------------------------------------------------------

print("\n--- 2.5 Тепловая карта ---")

plt.figure(figsize=(8, 6))
corr_matrix = skaters[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Purples', center=0,
            square=True, linewidths=1, fmt='.2f')
plt.title('Корреляция числовых признаков', fontsize=14)
plt.tight_layout()
plt.savefig('figures/3_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

print("Вывод: goals и assists сильно коррелируют (0.8+). points предсказуемо коррелирует с обоими.")


# -----------------------------------------------------------------------------
# 2.6 УДАЛЕНИЕ ДУБЛИКАТОВ
# -----------------------------------------------------------------------------

print("\n--- 2.6 Удаление дубликатов ---")

duplicates = skaters.duplicated().sum()
print(f"Найдено дубликатов: {duplicates}")
if duplicates > 0:
    skaters = skaters.drop_duplicates()
    print("Дубликаты удалены")


# -----------------------------------------------------------------------------
# 2.7 АНАЛИЗ ВЫБРОСОВ (ящики с усами)
# -----------------------------------------------------------------------------

print("\n--- 2.7 Анализ выбросов ---")

fig, axes = plt.subplots(1, 4, figsize=(16, 5))

for i, col in enumerate(numeric_cols):
    axes[i].boxplot(skaters[col], patch_artist=True,
                    boxprops=dict(facecolor=LIGHT_PURPLE, color=DARK_PURPLE),
                    whiskerprops=dict(color=DARK_PURPLE),
                    capprops=dict(color=DARK_PURPLE))
    axes[i].set_title(f'Выбросы: {col}')
    axes[i].set_ylabel(col)

    # Считаем выбросы
    Q1 = skaters[col].quantile(0.25)
    Q3 = skaters[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = skaters[(skaters[col] < Q1 - 1.5*IQR) | (skaters[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} выбросов ({len(outliers)/len(skaters)*100:.1f}%)")

plt.tight_layout()
plt.savefig('figures/4_boxplots.png', dpi=150, bbox_inches='tight')
plt.show()

print("Вывод: Выбросы — это игроки-звезды, их не удаляем, они ценны для анализа.")


# -----------------------------------------------------------------------------
# 2.8 УСЛОВНАЯ ФИЛЬТРАЦИЯ
# -----------------------------------------------------------------------------

print("\n--- 2.8 Условная фильтрация ---")

# Фильтр 1: звезды (>500 очков)
stars = skaters[skaters['points'] > 500]
print(f"\nФильтр 1 (звезды, >500 очков): {len(stars)} игроков")

# Фильтр 2: ветераны (>1000 игр)
veterans = skaters[skaters['games'] > 1000]
print(f"Фильтр 2 (ветераны, >1000 игр): {len(veterans)} игроков")

# Фильтр 3: снайперы (>300 голов)
snipers = skaters[skaters['goals'] > 300]
print(f"Фильтр 3 (снайперы, >300 голов): {len(snipers)} игроков")


# -----------------------------------------------------------------------------
# 2.9 ДОБАВЛЕНИЕ ШУМА
# -----------------------------------------------------------------------------

print("\n--- 2.9 Добавление шума ---")

# Добавляем шум в голы и передачи (5% от стандартного отклонения)
noise_goals = np.random.normal(0, skaters['goals'].std() * 0.05, len(skaters))
skaters['goals_noisy'] = skaters['goals'] + noise_goals

noise_assists = np.random.normal(0, skaters['assists'].std() * 0.05, len(skaters))
skaters['assists_noisy'] = skaters['assists'] + noise_assists

# Визуализируем эффект шума
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(skaters['goals'], bins=40, alpha=0.5, label='Исходные', color=PURPLE)
axes[0].hist(skaters['goals_noisy'], bins=40, alpha=0.5, label='С шумом', color=LIGHT_PURPLE)
axes[0].set_title('Голы: исходные vs с шумом')
axes[0].legend()

axes[1].hist(skaters['assists'], bins=40, alpha=0.5, label='Исходные', color=PURPLE)
axes[1].hist(skaters['assists_noisy'], bins=40, alpha=0.5, label='С шумом', color=LIGHT_PURPLE)
axes[1].set_title('Передачи: исходные vs с шумом')
axes[1].legend()

plt.tight_layout()
plt.savefig('figures/5_noise_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("Шум добавлен для устойчивости модели к погрешностям измерений.")


# -----------------------------------------------------------------------------
# 2.10 ПРЕОБРАЗОВАНИЕ ЧИСЛОВЫХ ДАННЫХ В КАТЕГОРИАЛЬНЫЕ
# -----------------------------------------------------------------------------

print("\n--- 2.10 Числовые → категориальные ---")

# Разбиваем игры на категории
skaters['games_category'] = pd.cut(skaters['games'],
                                   bins=[0, 100, 500, 1000, float('inf')],
                                   labels=['Новичок (0-100)', 'Опытный (101-500)',
                                           'Ветеран (501-1000)', 'Легенда (1000+)'])

print("Распределение по категориям игр:")
print(skaters['games_category'].value_counts())

# Визуализация
plt.figure(figsize=(10, 5))
skaters['games_category'].value_counts().plot(kind='bar', color=PURPLE, edgecolor=DARK_PURPLE)
plt.title('Категории игроков по количеству игр')
plt.xlabel('Категория')
plt.ylabel('Количество игроков')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('figures/6_games_categories.png', dpi=150, bbox_inches='tight')
plt.show()


# -----------------------------------------------------------------------------
# 2.11 ДОПОЛНИТЕЛЬНЫЕ ПРЕОБРАЗОВАНИЯ
# -----------------------------------------------------------------------------

print("\n--- 2.11 Дополнительные преобразования ---")

# Унифицируем позиции (они уже унифицированы, просто выводим)
if 'position' in skaters.columns:
    print(f"Уникальные позиции: {skaters['position'].unique()}")
    print("Позиции уже унифицированы.")


# -----------------------------------------------------------------------------
# 2.12 ОЦЕНКА ИЗМЕНЕНИЙ ПОСЛЕ ФИЛЬТРАЦИИ
# -----------------------------------------------------------------------------

print("\n--- 2.12 Оценка изменений после фильтрации ---")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Все игроки
axes[0].hist(skaters['points'], bins=40, color=PURPLE, alpha=0.7, edgecolor=DARK_PURPLE)
axes[0].set_title('Все игроки')
axes[0].set_xlabel('Очки')
axes[0].set_ylabel('Количество')

# Только звезды (>500 очков)
axes[1].hist(stars['points'], bins=40, color=LIGHT_PURPLE, alpha=0.7, edgecolor=DARK_PURPLE)
axes[1].set_title('Звезды (>500 очков)')
axes[1].set_xlabel('Очки')
axes[1].set_ylabel('Количество')

plt.tight_layout()
plt.savefig('figures/7_filter_impact.png', dpi=150, bbox_inches='tight')
plt.show()

print("После фильтрации остаются только игроки с высокими показателями.")


# -----------------------------------------------------------------------------
# 2.13 ГРУППИРОВКА ДАННЫХ (объединение полевых игроков и вратарей)
# -----------------------------------------------------------------------------

print("\n--- 2.13 Группировка данных ---")

# Добавляем тип игрока
skaters['player_type'] = 'skater'
goalies['player_type'] = 'goalie'

# Объединяем для сравнения
skaters_subset = skaters[['games', 'points', 'player_type']].copy()
goalies_subset = goalies[['games', 'points']].copy()
goalies_subset['player_type'] = 'goalie'
all_players = pd.concat([skaters_subset, goalies_subset])

# Группировка
grouped = all_players.groupby('player_type').agg({
    'games': ['mean', 'count'],
    'points': ['mean']
}).round(2)
print(grouped)

# Визуализация сравнения
plt.figure(figsize=(8, 6))
means = [skaters['points'].mean(), goalies['points'].mean()]
bars = plt.bar(['Полевые игроки', 'Вратари'], means, color=[PURPLE, LIGHT_PURPLE])
plt.title('Среднее количество очков: полевые игроки vs вратари')
plt.ylabel('Среднее количество очков')

for bar, val in zip(bars, means):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             f'{val:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('figures/8_skaters_vs_goalies.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Вывод: Полевые игроки набирают в среднем {means[0]:.1f} очков, вратари — {means[1]:.1f}.")


# =============================================================================
# 3. ХАРАКТЕРИСТИКА КАТЕГОРИАЛЬНЫХ ДАННЫХ
# =============================================================================

print("\n" + "=" * 60)
print("3. ХАРАКТЕРИСТИКА КАТЕГОРИАЛЬНЫХ ДАННЫХ")
print("=" * 60)

# -----------------------------------------------------------------------------
# 3.1 ПЕРЕЧЕНЬ КАТЕГОРИЙ
# -----------------------------------------------------------------------------

print("\n--- 3.1 Категории ---")

if 'position' in skaters.columns:
    print(f"Позиции: {skaters['position'].unique()}")
if 'active' in skaters.columns:
    print(f"Активность: активны {skaters['active'].sum()} игроков, неактивны {len(skaters) - skaters['active'].sum()}")
if 'hof' in skaters.columns:
    print(f"Зал славы: {skaters['hof'].sum()} игроков")


# -----------------------------------------------------------------------------
# 3.2 ДИАГРАММЫ РАСПРЕДЕЛЕНИЯ КАТЕГОРИЙ
# -----------------------------------------------------------------------------

print("\n--- 3.2 Диаграммы категорий ---")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Распределение по позициям
if 'position' in skaters.columns:
    position_counts = skaters['position'].value_counts()
    axes[0].bar(position_counts.index, position_counts.values, color=PURPLE, edgecolor=DARK_PURPLE)
    axes[0].set_title('Распределение по позициям')
    axes[0].set_xlabel('Позиция')
    axes[0].set_ylabel('Количество игроков')
    axes[0].tick_params(axis='x', rotation=45)

# Активность
if 'active' in skaters.columns:
    active_counts = skaters['active'].value_counts()
    axes[1].pie(active_counts.values, labels=['Неактивные', 'Активные'],
                autopct='%1.1f%%', colors=[LIGHT_PURPLE, PURPLE])
    axes[1].set_title('Активные vs Неактивные игроки')

plt.tight_layout()
plt.savefig('figures/9_categorical.png', dpi=150, bbox_inches='tight')
plt.show()


# -----------------------------------------------------------------------------
# 3.3 ПРЕОБРАЗОВАНИЕ КАТЕГОРИЙ В ЧИСЛА
# -----------------------------------------------------------------------------

print("\n--- 3.3 Преобразование в числовые ---")

# One-Hot Encoding для позиций
if 'position' in skaters.columns:
    position_encoded = pd.get_dummies(skaters['position'], prefix='pos')
    print(f"One-Hot Encoding: добавлено {position_encoded.shape[1]} новых признаков")
    print(position_encoded.head())

# Label Encoding для активности
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
skaters['active_encoded'] = le.fit_transform(skaters['active'])
print(f"\nLabel Encoding активности: 'False' → {le.transform(['False'])[0]}, 'True' → {le.transform(['True'])[0]}")


# =============================================================================
# 4. ГИПОТЕЗЫ И РИСКИ
# =============================================================================

print("\n" + "=" * 60)
print("4. ГИПОТЕЗЫ И РИСКИ")
print("=" * 60)

print("\n--- ГИПОТЕЗЫ ДЛЯ ИСПОЛЬЗОВАНИЯ ---")
print("""
1. Игроки с большим количеством игр набирают больше очков (опыт → результативность)
2. Существует связь между позицией и типом результативности (защитники больше передают)
""")

print("\n--- ГИПОТЕЗЫ ДЛЯ НЕЭТИЧНОГО ИСПОЛЬЗОВАНИЯ ---")
print("""
1. Прогнозирование "сливов" матчей для манипуляций на тотализаторе
2. Дискриминация молодых игроков при найме на основе статистических прогнозов
""")

print("\n--- ВОЗМОЖНОСТИ РАСШИРЕНИЯ ---")
print("""
Можно добавить:
- Данные о зарплатах (экономический анализ)
- Данные о травмах (прогноз доступности)
- Физические параметры (рост, вес)
- Командную статистику
""")


# =============================================================================
# ВЫВОДЫ
# =============================================================================

print("\n" + "=" * 60)
print("ВЫВОДЫ ПО ПЕРВОМУ РАЗДЕЛУ")
print("=" * 60)

print("""
1. ХАРАКТЕРИСТИКА НАБОРА:
   - 7,785 полевых игроков, 869 вратарей
   - Период: 1918-2025 гг.
   - Признаки: игры, голы, передачи, очки, позиция, активность

2. ВЫЯВЛЕННЫЕ ПРОБЛЕМЫ:
   - Пропуски в 'time' (45.8%) — заполнены средним
   - Выбросы (игроки-звезды) — оставлены для анализа
   - Асимметрия распределений

3. ПРИГОДНОСТЬ ДАННЫХ:
   - Датасет пригоден для задач классификации и регрессии
   - Требуется нормализация и кодирование категорий

4. РЕКОМЕНДАЦИИ:
   - Нормализовать числовые признаки
   - Применить One-Hot Encoding для позиций
   - Использовать points как целевую переменную
""")

print("\n" + "=" * 60)
print("АНАЛИЗ ЗАВЕРШЕН")
print("=" * 60)
```
